```python
import pandas as pd
```

```python
df = pd.read_excel("/Net/Groups/BSI/work_scratch/ppapastefanou/data/Fluxnet_detail/raw/psi_midday_all.xlsx")
df["date"] = pd.to_datetime(df["date"])
```

```python
df
```

```python
df['psi_leaf_midday_avg'] = df[['psi_1', 'psi_2']].mean(axis=1)
df['prefix'] = df['sample_id'].str.split('_').str[0]
df_clean = df[df['prefix'] == 'FASY'].groupby(['date', 'prefix'])['psi_leaf_midday_avg'].mean().reset_index()
# Convert from bar to MPa
df_clean['psi_leaf_midday_avg'] *= -0.1

df_clean['date'] = df_clean['date'].dt.normalize() + pd.Timedelta(hours=14)
df_clean = df_clean[(df_clean['date'] <= '2024-09-11')]
df_clean
```

```python
df_clean.to_csv("/Net/Groups/BSI/work_scratch/ppapastefanou/data/Fluxnet_detail/eval_processed/psi_leaf_midday_2023_2024_avg.csv")
```
